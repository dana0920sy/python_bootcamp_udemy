```python
# Global Variables will also be needed that you can pass in the instances
# Dictionary of values and map to the values
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs') #Tuples so that they are immutable
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace') #Tuples so that they are immutable
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card():
    #Define the instance/construct with the attributes and parameters that need to be passed in
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    #Methods
    def __str__(self):
        return self.rank+ " of "+self.suit
```

```python
# Deck Class
# Purposes is to instantiate a NEW DECK
# Idea is to create a deck of 52 cxard objects
# hold the list of the card objects
# Shuffle - use of random library shuffle function
# Deal cards

# Introduction to holding of object instances
class Deck():
    
    def __init__(self):
        # Create an instance of the deck
        self.all_cards = [] # Empty list with no input from the user
        
        #Need to create a card object - two iterations in one
        for suit in suits:
            for rank in ranks:
                #Creation of card instances and appending it to the cards list
                created_card = Card(suit,rank)
                self.all_cards.append(created_card)
    
    # Shuffle Method
    def shuffle(self):
        random.shuffle(self.all_cards) # This does not return anything
    
    # Deal Method
    
    def deal_one(self):
        return self.all_cards.pop()
```

```python
# Player Class
# Holds current player's cards
# Player should be able to add or remove cards from their hand (their list of card objects)
# Add a single card or multiple cards in their list
# Translate deck with a top and bottom to a Python list

#extend() - adding multiple cards to bottom of the list

class Player():
    
    def __init__(self,name):
        self.name = name
        self.all_cards=[]
    # Method for removing card
    def remove_one(self):
        return self.all_cards.pop(0)
    
    #Method for adding cards to hand
    def add_cards(self,new_cards):
        # List of multiple card objects
        if type(new_cards)==type([]):
            self.all_cards.extend(new_cards)
        # Single card object
        else:
            self.all_cards.append(new_cards)

    def __str__(self):
        return f"Player {self.name} has {len(self.all_cards)} cards."        
```

```python
# Game Logic
# Create two instances of P1 and P2
player_one = Player("P1")
player_two = Player("P2")
```

```python
# Create a new deck
new_deck = Deck()
```

```python
#Shuffle the deck
new_deck.shuffle()
```

```python
len(new_deck.all_cards)/2
```

```python
#Split the deck between the two players
for i in range(26):
    player_one.add_cards(new_deck.deal_one())
    player_two.add_cards(new_deck.deal_one())
```

```python
# Game Proper
round_counter = 0
game_on = True

while game_on:
    
    round_counter+=1
     # Check if either player has 0 cards to play
    print(f"Round {round_counter}")
    if len(player_one.all_cards)==0:
        print("Player 1 has no more cards. Player 2 wins!")
        game_on = False
        break
    if len(player_two.all_cards)==0:
        print("Player 2 has no more cards. Player 1 wins!")
        game_on=False
        break

    #Play round
    player_one_cards=[]
    player_one_cards.append(player_one.remove_one())
    player_two_cards=[]
    player_two_cards.append(player_two.remove_one())
    
    #Let at_war = True
    
    at_war = True
    
    while at_war:
        
        #Scenario if player 1 wins
        
        if player_one_cards[-1].value > player_two_cards[-1].value:
            player_one.add_cards(player_one_cards)
            player_one.add_cards(player_two_cards)
            at_war = False
        #Scenario if player 2 wins
        elif player_one_cards[-1].value < player_two_cards[-1].value:
            player_two.add_cards(player_one_cards)
            player_two.add_cards(player_two_cards)
            at_war = False
        
        #If there's a draw
        else:
            print("war!")
            # War!
            # Keep on drawing cards
            # if Player 1 has less cards to play / draw
            if len(player_one.all_cards)<3:
                print("Player 1 has no more cards to deal with. Player 2 wins!")
                game_on=False
                break
            elif len(player_two.all_cards)<3:
                print("Player 2 has no more cards to deal with. Player 1  wins!")
                game_on = False
                break
            else:
                for num in range(3):
                    player_one_cards.append(player_one.remove_one())
                    player_two_cards.append(player_two.remove_one())
   
```
