```python
import requests
import bs4
```

```python
base_url = requests.get('http://quotes.toscrape.com/')
result = bs4.BeautifulSoup(base_url.text,'lxml')
result
```

```python
# URL Structure for each page https://quotes.toscrape.com/page/1/

# Let's get the first page

first_page_url = requests.get('https://quotes.toscrape.com/page/1/')
result_first_page = bs4.BeautifulSoup(first_page_url.text,'lxml')
quote_block = result_first_page.select('.quote')

set_authors = set([quote_block[i].select('.author')[0].text for i in range(len(quote_block))])


set_authors
```

```python
# List of quotes on the first page

list_quotes = [quote_block[i].select('.text')[0].text for i in range(len(quote_block))]
list_quotes
```

```python
# Inspect site to extract top 10 tags

top_ten_tags = result_first_page.select('.tag-item')
for i in range(len(top_ten_tags)):
    print(top_ten_tags[i].text)
```

```python
base_url = requests.get('http://quotes.toscrape.com/page/100')
result = bs4.BeautifulSoup(base_url.text,'lxml')
"No quotes found!" in result.text


keep_check = True
page_num = 1
authors = set()
while keep_check:
    
    url = 'http://quotes.toscrape.com/page/{}'
    page_url = requests.get(url.format(page_num))
    result = bs4.BeautifulSoup(page_url.text,'lxml')
    
    if 'No quotes found!' in result.text:
        keep_check = False
    else:
        quote_block = result.select('.quote')
        for i in range(len(quote_block)):
            author = quote_block[i].select('.author')[0].text
            authors.add(author)
    page_num +=1

authors
```
